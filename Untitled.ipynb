```python
import pandas as pd
```

```python
millionDF= pd.read_csv("SimpleHandValue_v_SimpleHandValue.csv")
```

```python
millionDF.head()
```

```python
millionDF.describe()
```

```python
for col in millionDF.columns:
    millionDF[col]=millionDF[col].astype(int)
```

```python
millionDF.describe()
```

```python
millionDF.count()
```

```python
millionDF['Count']=millionDF.index+1
millionDF['SimpleHandValue_ratio']=millionDF['SimpleHandValue']/millionDF['Count']
millionDF['SimpleHandValue2_ratio']=millionDF['SimpleHandValue2']/millionDF['Count']
millionDF['Average']=(millionDF['SimpleHandValue_ratio']+millionDF['SimpleHandValue2_ratio'])/2
millionDF['Ties_ratio']=millionDF['Ties']/millionDF['Count']
```
